```python
#Turotial 9 https://www.youtube.com/watch?v=7DK70jLZBzY&index=12&list=PLQVvvaa0QuDfefDfXb9Yf0la1fPDKluPF import datetime as dt

from matplotlib import style

theDate = ['20170101', '20170102', '20170103', '20170104','20170105','20170106','20170107','20170108','20170109','20170110']

# ggplot is one of the style
style.use('ggplot')
style.use('fivethirtyeight')


formatedDate = [dt.datetime.strptime(d,'%Y%m%d').date() for d in theDate] 
theValue = range(10)

fig = plt.figure() 
ax1 = plt.subplot2grid((1,1), (0,0))
#- means use line rather than dot
#plt.plot_date(formatedDate, theValue, '*') 
ax1.plot_date(formatedDate, theValue, '-', label='stuff') 
plt.xlabel('Date') 
plt.ylabel('Price')
ax1.xaxis.label.set_color('g')
ax1.yaxis.label.set_color('r')
ax1.spines['left'].set_color('g') 
ax1.spines['right'].set_visible(False) 
ax1.spines['top'].set_visible(False)

#change the width of the left 
ax1.spines['left'].set_linewidth(5)
#change the x tick color
ax1.tick_params(axis='x', colors='#f06215')
ax1.set_yticks([0, 3, 6, 9])

# 

#set an obvious base line
baseline = 3
ax1.axhline(baseline, color='r', linewidth=1)
#the third parameter decide where is the baseline being draw
ax1.fill_between(formatedDate, theValue, baseline, alpha=0.1)
plt.title('graphs') 
plt.legend()

for label in ax1.xaxis.get_ticklabels(): label.set_rotation(45)
#plt.show()
```

```python
#tutorial 16, live graphs
import matplotlib.animation as animation
from matplotlib import style

style.use('fivethirtyeight')

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

def animate(i):
    graph_date = open('/home/edwinguo/rbc/example.txt', 'r').read()
    lines = graph_date.split('\n')
    xs = []
    ys = []
    for line in lines:
        if len(line) > 1: 
            x,y = line.split(',')
            xs.append(x)
            ys.append(y)
    ax1.clear()
    ax1.plot(xs,ys)

ani = animation.FuncAnimation(fig, animate, interval=1000)
    
plt.show()
```

```python
#subplot
import matplotlib.pyplot as plt
import random
from matplotlib import style

style.use('fivethirtyeight')

fig = plt.figure()

def create_plots():
    xs = []
    ys = []
    for i in range(10):
        x = i
        y = random.randrange(10)
        
        xs.append(x)
        ys.append(y)
    return xs, ys

ax1 = plt.subplot2grid((6,1), (0,0), rowspan = 2, colspan = 2)   
ax2 = plt.subplot2grid((6,1), (2,0), rowspan = 2, colspan = 3)
ax3 = plt.subplot2grid((6,1), (4,0), rowspan = 2, colspan = 3)
#fig.add_subplot(223)   #bottom left
#fig.add_subplot(224)   #bottom right 

# subplot(2,2,1:2)
# subplot(2,2,3)
# subplot(2,2,4)

x1, y1 = create_plots()
ax1.plot(x1, y1)

x2, y2 = create_plots()
ax2.plot(x2, y2)

x3, y3 = create_plots()
ax3.plot(x3, y3)

plt.show()
```
